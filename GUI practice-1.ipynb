```python
from tkinter import *
def changecolor(e):
    ch=e.char             # e.char is very important for the response of keys on keyboard
    if ch=="r":
        root.config(bg="red")
    elif ch=="b":
        root.config(bg="blue")
    elif ch=="g":
        root.config(bg="green")

root = Tk()
root.geometry('200x200')
root.bind("<Key>",changecolor)
root.mainloop()
```

```python
from tkinter import *

root = Tk()
root.geometry('300x300')



def finish():
    return root.destroy()


def divide():
    a=int(e1.get())
    b=int(e2.get())
    e3.insert(0,a/b)
 
    
def clear():
    
    e1.delete(0,END) #clears the entries from 0 to end
    e2.delete(0,END)
    e3.delete(0,END)
    e1.focus()
    
    
root.title('Number Circus')

l0=Label(root, text = " Division Alert", font='Ubuntu')
l1=Label(root, text="First Number")
l2=Label(root, text="Last Number")

e1 =Entry(root)
e2 =Entry(root)
e3 =Entry(root)

b1=Button(root, text='Divide', command=divide)
b2=Button(root, text='Clear', command=clear)
b3=Button(root,text='Close',command=finish)

l0.grid(row=0,column=0)
l1.grid(row=1,column=0)
l2.grid(row=2, column=0)
e1.grid(row=1, column=1)
e2.grid(row=2, column =1)
e3.grid(row=3,column=1)
b1.grid(row=4, column=0, sticky=W, pady=4)
b2.grid(row=4, column=1, sticky=W, pady=4)
b3.grid(row=4, column=2, sticky=W, pady=4)


root.mainloop( )
```

```python
from tkinter import *

def divide():
    try:
        e3.delete(0, END)
        e3.config(fg="red")
        a=int(e1.get())
        b=int(e2.get())
        c=a/b
        e3.config(fg="green")
        e3.insert(0,str(c))
    except ZeroDivisionError:
        e3.insert(0,"Can't Divide by 0!")
    except ValueError:
        e3.insert(0,"Only digits allowed!")

def clear():
    e1.delete(0, END)
    e2.delete(0, END)
    e3.delete(0, END)
    e1.focus()

def finish():
    root.destroy()
    
root = Tk()
root.geometry('400x200+100+200')
l1=Label(root, text="Number Division Program",font="Arial 18 bold")
l2=Label(root, text="First No:")
l3=Label(root, text="Second No:")
l4=Label(root,text="Result:")
e1=Entry(root)
e2=Entry(root)
e3=Entry(root)
b1=Button(root,text="Divide",command=divide)
b2=Button(root,text="Clear",command=clear)
b3=Button(root,text="Quit",command=finish)
l1.grid(row=0,column=0,columnspan=4)
l2.grid(row=1,column=0,sticky=E)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0,sticky=E)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0,sticky=E)
e3.grid(row=3,column=1)
b1.grid(row=4,column=0,pady=5,padx=5,sticky=E+W)
b2.grid(row=4,column=1,pady=5,padx=5,sticky=E+W)
b3.grid(row=4,column=2,pady=5,padx=5,sticky=E+W)
root.mainloop()
    
    
    
    
```
